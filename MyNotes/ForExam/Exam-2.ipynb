```python
# problem 2.2.2
def normpdf(x,mu,sigma):       
    pdf=np.exp(-(x-mu)**2/(2*sigma**2))/(sigma * np.sqrt(2 * np.pi))
    return pdf

mu1,sigma1=39.133,13.207 # drug group
x1= np.arange(mu1-3*sigma1,mu1+3*sigma1,0.01) 
y1=normpdf(x1,mu1,sigma1)

mu2,sigma2=45.9,6.136 # placebo group
x2= np.arange(mu2-3*sigma2,mu2+3*sigma2,0.01)
y2=normpdf(x2,mu2,sigma2)

fig, ax = plt.subplots(figsize=(12, 6)) 
ax.set(xlabel="size", ylabel="Frequency")
ax.plot(x1,y1, 'red', linewidth=2, label='group 1')   
ax.plot(x2,y2, 'purple', linewidth=2, label='group 2') 
plt.title('distribution of two groups')

d = {
    'mu1': 39.133 ,
    'sigma1':13.207,
    'mu2': 45.9 ,
    'sigma2':6.136,

    }
    

text = nice_string_output(d, extra_spacing=2, decimals=3)
add_text_to_ax(0.75, 0.95, text, ax, fontsize=20)

ax.legend(loc='upper left', fontsize=18); 

fig.tight_layout()
plt.savefig('fig7',dpi=600)

plt.show()
```

```python
# problem 2.2.2
def solve(m1,m2,std1,std2):
    a = 1/(2*std1**2) - 1/(2*std2**2)
    b = m2/(std2**2) - m1/(std1**2)
    c = m1**2 /(2*std1**2) - m2**2 / (2*std2**2) - np.log(std2/std1)
    return np.roots([a,b,c])

m1 = 39.133
std1 = 13.207
m2 = 45.9
std2 = 6.136

result = solve(m1,m2,std1,std2)
print(result)
```
